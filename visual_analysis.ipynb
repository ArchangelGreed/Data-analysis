```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
games  = 'games.csv'
chess_data = pd.read_csv(games,index_col='id')
```

```python
chess_data.columns
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x=chess_data['winner'],y=chess_data['turns'])
plt.title('The Average No Of Moves For Each Player')
```

```python
chess_data.head()
```

```python
plt.figure(figsize=(15,6))

sns.scatterplot(x=chess_data['winner'],y=chess_data['turns'])
```

```python
sns.scatterplot(x=chess_data['white_rating'],y=chess_data['turns'],hue=chess_data['winner'])
```

```python
plt.figure(figsize=(15,6))
sns.regplot(x=chess_data['black_rating'],y=chess_data['turns'])
```

```python
sns.scatterplot(x=chess_data['white_rating'],y=chess_data['black_rating'],hue=chess_data['victory_status'])
plt.savefig('chess_scatterplot')

```

```python
plt.figure(figsize=(20,6))
sns.barplot(chess_data['winner'])
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(20,6))
sns.pairplot(chess_data)
```
